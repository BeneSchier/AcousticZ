# Examples for using AcousticZ Python Package

## Example 1: Shoexbox

### Importing the package

```python
import numpy as np
from AcousticZ.Room import Room
```

### Loading the .obj mesh

```python
# Defining path to .obj mesh
shoebox_file = '../data/example_meshes/shoebox.obj'
shoebox = Room(shoebox_file)
```

### Add source and receiver to the room

```python
# Define source and receiver coordinates
source = np.array([2.0, 2.0, 2.0])
receiver = np.array([5.0, 5.0, 1.8])

# Adding source 
shoebox.createSource(source)

# Addding receiver
shoebox.createReceiver(receiver, radiusOfReceiverSphere=0.0875)

shoebox.showRoom()
```

### Perform the Ray Tracing procedure

```python
# Perform Ray Tracing with 5000 rays
shoebox.performRayTracing(5000)
```

### RIR Generation

```python
# Generate the Room Impulse Response
shoebox.generateRIR()
shoebox.plotWaveform()
```

### Applying RIR to audio file

```python
# Define path to audio file
audio_file = '../data/example_audio/drums.wav'
# Define output path
output_path = './out/processed_audio_shoebox.wav'
shoebox.applyRIR(audio_file, output_path)
```

## Example 2: Living Room

### Loading the .obj mesh

```python
# Defining path to .obj mesh
room_file = '../data/example_meshes/InteriorTest.obj'
room = Room(room_file)
```

### Add source and receiver to the room

```python
# Define source and receiver coordinates
source = np.array([2.0, 2.0, 2.0])
receiver = np.array([2.0, 2.0, 1.8])

# Adding source 
room.createSource(source)

# Addding receiver
room.createReceiver(receiver, radiusOfReceiverSphere=0.0875)

room.showRoom()
```

### Perform the Ray Tracing procedure

```python
# Perform Ray Tracing with 5000 rays
room.performRayTracing(5000)
```

### RIR Generation

```python
# Generate the Room Impulse Response
room.generateRIR()
room.plotWaveform()
```

```python
audio_file = '../data/example_audio/drums.wav'
output_path = './out/processed_audio_livingRoom.wav'
room.applyRIR(audio_file, output_path)
```
